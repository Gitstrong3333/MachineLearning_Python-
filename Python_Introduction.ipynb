```python
Python is famous for its clean syntax and its strict rules on indentation.

üêç Python Syntax Basics
‚úÖ General Rules

Case-sensitive

print is valid, but Print or PRINT is not.

Variables name and Name are different.

Statements end with a newline (no ; needed).

x = 5
y = 10
print(x + y)


Comments start with #

# This is a comment
print("Hello")


Multi-line statements can be split using \ or inside ()

total = 1 + 2 + 3 + \
        4 + 5


or

total = (1 + 2 + 3 +
         4 + 5)

üêç Python Indentation Rules

Unlike many languages (like C, Java, JavaScript) that use { } to define code blocks, Python uses indentation (spaces or tabs).

üëâ Indentation is mandatory in Python.

‚úÖ Example:
if True:
    print("This is indented correctly")
    print("Still inside the if-block")
print("Outside the if-block")

‚ùå Wrong (will throw an IndentationError):
if True:
print("This will cause an error")

üìù Indentation Guidelines

Use 4 spaces per indentation level (PEP 8 recommendation).

Do not mix tabs and spaces ‚Üí causes errors.

Blocks that belong together must have the same indentation level.

üìå Rules for Writing Python Scripts

File should be saved with .py extension.

Start with imports at the top:

import math
import sys


Variables and functions should follow snake_case:

student_name = "Alice"
def calculate_total(a, b):
    return a + b


Constants should be in UPPER_CASE:

PI = 3.14159


Indentation must be consistent throughout the file.

Use blank lines to separate functions and classes for readability.

‚úÖ Example Full Script (hello.py):

# This is my first Python script

# Function to greet
def greet(name):
    print(f"Hello, {name}!")

# Main execution
if __name__ == "__main__":
    greet("SSSS")
    print("Script executed successfully!")


Run it:

python hello.py
```

```python
Python comments in detail ‚Äî they are super important for writing clean, understandable code.

üêç Python Comments
‚úÖ What are comments?

Notes in your code that Python ignores during execution.

Used to explain logic, improve readability, or temporarily disable code.

1. Single-line Comments

Start with # ‚Üí everything after it on the same line is ignored by Python.

# This is a single-line comment
x = 5  # You can also put a comment at the end of a line
print(x)  # This prints the value of x

2. Multi-line Comments (Block Comments)

Python doesn‚Äôt have a dedicated multi-line comment syntax (like /* ... */ in C).
But you can:

Use # on each line

Or use triple quotes """ ... """ or ''' ... ''' (commonly for docstrings but can be used as block comments).

# This is a block comment
# spread across multiple lines
# explaining the logic

x = 10
y = 20
print(x + y)


Or:

"""
This is also a comment
spanning multiple lines.
It will not be executed.
"""
print("Hello, World!")

3. Docstrings (Documentation Strings)

Special type of comment (technically a string literal) used inside functions, classes, or modules to describe their purpose.
They are written with triple quotes """ ... """.

def greet(name):
    """
    This function greets the person passed as argument.
    :param name: string
    :return: None
    """
    print(f"Hello, {name}!")


üëâ Docstrings can be accessed at runtime using:

print(greet.__doc__)

4. Commenting Out Code (Temporary Disable)

You can use # to prevent a line from running:

# print("This line won't run")
print("This line will run")

üìå Best Practices for Comments

Write comments that explain why, not just what.

# Bad: Increment x by 1
x = x + 1

# Good: Increment x to move to the next index in the list
x = x + 1


Keep comments concise and clear.

Update comments when you change the code.

Use docstrings for functions, classes, and modules.

‚úÖ Example Script with Comments:

# Program to add two numbers

def add_numbers(a, b):
    """Returns the sum of two numbers a and b."""
    return a + b

# Main program
if __name__ == "__main__":
    x = 10  # first number
    y = 20  # second number
    result = add_numbers(x, y)
    print("The sum is:", result)  # Display the result
```

```python
Python Variables step by step.

üêç Python Variables
‚úÖ What is a Variable?

A named storage that holds data (a value) in memory.

In Python, you don‚Äôt need to declare a type (like in C/Java).

Type is decided at runtime ‚Üí Python is dynamically typed.

x = 10        # integer
name = "John" # string
pi = 3.14     # float
is_active = True  # boolean

üìå Rules for Naming Variables

Must start with a letter or underscore _.
‚úÖ my_var, _temp
‚ùå 2value, @name

Can only contain letters, numbers, and underscores.
‚úÖ total_sum1, user_age

Case-sensitive ‚Üí age, Age, and AGE are different.

Should not use Python keywords (if, class, while, etc.).

üìå Variable Assignment

Simple assignment:

x = 5
y = "Hello"


Multiple assignments:

a, b, c = 10, 20, 30
print(a, b, c)  # 10 20 30


Same value to multiple variables:

x = y = z = 100
print(x, y, z)  # 100 100 100

üìå Variable Types

Python decides type automatically (dynamic typing):

num = 42        # int
price = 19.99   # float
name = "Alice"  # str
is_valid = True # bool


Check type:

print(type(num))   # <class 'int'>
print(type(name))  # <class 'str'>

üìå Constants

Python doesn‚Äôt have true constants, but by convention:

Use UPPERCASE names to indicate constant values.

PI = 3.14159
MAX_USERS = 100

üìå Variable Scope

Global variable ‚Üí defined outside functions, accessible everywhere.

Local variable ‚Üí defined inside a function, only accessible there.

x = 10  # global

def test():
    y = 5  # local
    print("Inside:", y)

test()
print("Outside:", x)

‚úÖ Example Full Script
# Example showing variables

# Global variable
app_name = "MyPythonApp"

def greet(user):
    """Function to greet the user."""
    greeting = f"Hello, {user}!"  # local variable
    return greeting

# Assign values
age = 25
height = 5.7
is_student = False

# Output
print(greet("Sharmila"))
print("Age:", age)
print("Height:", height)
print("Student:", is_student)
```

```python
#Python Casting
Let‚Äôs go into Python Casting (Type Conversion).

In Python, casting means converting one data type into another.

üîπ 1. Casting to int

Converts a number or string to an integer (whole number).

x = int(3.8)     # float ‚Üí int ‚Üí 3
y = int("15")    # string ‚Üí int ‚Üí 15
print(x, y)      # 3 15


‚ö†Ô∏è Note: decimals are truncated (not rounded).

üîπ 2. Casting to float

Converts to decimal number.

a = float(5)     # int ‚Üí float ‚Üí 5.0
b = float("3.14")# string ‚Üí float ‚Üí 3.14
print(a, b)      # 5.0 3.14

üîπ 3. Casting to str

Converts to string (text).

num = str(100)   # int ‚Üí str ‚Üí "100"
pi = str(3.14)   # float ‚Üí str ‚Üí "3.14"
print(num, pi)   # "100" "3.14"

üîπ 4. Casting to complex

Converts to a complex number.

z1 = complex(5)     # int ‚Üí complex ‚Üí (5+0j)
z2 = complex(3.2)   # float ‚Üí complex ‚Üí (3.2+0j)
z3 = complex(2, 3)  # two arguments ‚Üí (2+3j)
print(z1, z2, z3)   # (5+0j) (3.2+0j) (2+3j)
```

```python
#Python Strings
Let‚Äôs go through Python Strings step by step.

üîπ What is a String in Python?

A string is a sequence of characters enclosed in single quotes ('), double quotes ("), or triple quotes (''' or """).

#Strings are one of the most commonly used data types in Python.

üîπ Examples of Strings
# Using single quotes
str1 = 'Hello'

# Using double quotes
str2 = "Python"

# Using triple quotes (for multi-line text)
str3 = '''This is
a multi-line
string'''

print(str1)
print(str2)
print(str3)


‚úÖ Output:

Hello
Python
This is
a multi-line
string

üîπ Accessing Strings

Strings work like arrays of characters ‚Üí you can access them using an index.

text = "Python"
print(text[0])   # First character ‚Üí 'P'
print(text[1])   # Second character ‚Üí 'y'
print(text[-1])  # Last character ‚Üí 'n'

üîπ String Slicing

You can get parts of a string using [start:end].

word = "Programming"
print(word[0:6])   # 'Progra'
print(word[3:])    # 'gramming'
print(word[:5])    # 'Progr'

üîπ String Operations
# Concatenation
a = "Hello"
b = "World"
print(a + " " + b)   # Hello World

# Repetition
print(a * 3)         # HelloHelloHello

# Length of string
print(len(a))        # 5

üîπ String Methods (Commonly Used)
s = "  Python Programming  "

print(s.lower())        # '  python programming  '
print(s.upper())        # '  PYTHON PROGRAMMING  '
print(s.strip())        # 'Python Programming' (removes spaces)
print(s.replace("Python", "Java"))  # '  Java Programming  '
print(s.split())        # ['Python', 'Programming']
print("Python" in s)    # True
```

```python
#Python Booleans
Let‚Äôs dive into Python Booleans step by step.

üîπ Python Booleans

Boolean values represent truth values:

True

False

In Python, Booleans are a data type just like integers or strings.

‚úÖ Basic Example
x = True
y = False

print(x)      # True
print(y)      # False
print(type(x))  # <class 'bool'>

üîπ Boolean as a Result of Expressions

Booleans often come from comparisons:

print(10 > 5)   # True
print(10 == 5)  # False
print(10 < 5)   # False

üîπ Using Booleans in Conditions
a = 15
if a > 10:
    print("a is greater than 10")
else:
    print("a is not greater than 10")


‚úÖ Output:

a is greater than 10

üîπ bool() Function

The bool() function converts a value into a Boolean (True or False).

print(bool(10))       # True
print(bool("Hello"))  # True
print(bool(""))       # False (empty string)
print(bool(0))        # False
print(bool([]))       # False (empty list)
print(bool(None))     # False


üìå Rule of Thumb:

Empty values ‚Üí False

Non-empty values ‚Üí True

üîπ Boolean Operators

Python has logical operators that work with Booleans:

x = True
y = False

print(x and y)  # False (both must be True)
print(x or y)   # True  (at least one True)
print(not x)    # False (negation)

üîπ Common Use Case in Programming

Booleans are heavily used in loops, conditions, and function checks:

name = "Python"
if name.startswith("Py"):
    print("Yes, it starts with 'Py'")
else:
    print("No")
```

```python
#Python Operators
Let‚Äôs cover Python Operators step by step.

üîπ Python Operators

Operators are symbols that perform operations on variables and values.

Example:

x = 10
y = 5
print(x + y)   # 15

‚úÖ 1. Arithmetic Operators

Used for math operations.

x = 10
y = 3

print(x + y)   # 13  (Addition)
print(x - y)   # 7   (Subtraction)
print(x * y)   # 30  (Multiplication)
print(x / y)   # 3.333... (Division - float)
print(x // y)  # 3   (Floor division - integer result)
print(x % y)   # 1   (Modulus - remainder)
print(x ** y)  # 1000 (Exponentiation: 10^3)

‚úÖ 2. Comparison Operators

Used to compare values (returns Boolean: True/False).

a = 10
b = 5

print(a == b)  # False (Equal to)
print(a != b)  # True  (Not equal to)
print(a > b)   # True
print(a < b)   # False
print(a >= b)  # True
print(a <= b)  # False

‚úÖ 3. Logical Operators

Used with Boolean values.

x = True
y = False

print(x and y)  # False
print(x or y)   # True
print(not x)    # False

‚úÖ 4. Assignment Operators

Used to assign values to variables (with shortcuts).

x = 10

x += 5   # x = x + 5 ‚Üí 15
x -= 3   # x = x - 3 ‚Üí 12
x *= 2   # x = x * 2 ‚Üí 24
x /= 4   # x = x / 4 ‚Üí 6.0
x %= 5   # x = x % 5 ‚Üí 1.0
x **= 3  # x = x ** 3 ‚Üí 1.0
x //= 2  # x = x // 2 ‚Üí 0.0

‚úÖ 5. Bitwise Operators

Work on binary (bits).

a = 6   # 110 in binary
b = 3   # 011 in binary

print(a & b)  # 2 ‚Üí 010 (AND)
print(a | b)  # 7 ‚Üí 111 (OR)
print(a ^ b)  # 5 ‚Üí 101 (XOR)
print(~a)     # -7 (NOT ‚Üí flips bits)
print(a << 1) # 12 ‚Üí 1100 (Left shift)
print(a >> 1) # 3  ‚Üí 011  (Right shift)

‚úÖ 6. Membership Operators

Check if a value is in a sequence.

numbers = [1, 2, 3, 4]

print(2 in numbers)     # True
print(10 not in numbers)  # True

‚úÖ 7. Identity Operators

Check if two objects are the same in memory.

x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is y)      # False (different objects with same values)
print(x is z)      # True (same object in memory)
print(x is not y)  # True


üìå Summary of Python Operators:

1.Arithmetic ‚Üí + - * / % // **

2.Comparison ‚Üí == != > < >= <=

3.Logical ‚Üí and or not

4.Assignment ‚Üí = += -= *= /= %= **= //=

5.Bitwise ‚Üí & | ^ ~ << >>

6.Membership ‚Üí in, not in

7.Identity ‚Üí is, is not
```

```python
#Python Lists
Let‚Äôs go step by step and learn about Python Lists.

üîπ What is a List?

A list in Python is a collection of items (elements) that are:

Ordered (elements have a defined order).

Mutable (can be changed: add, remove, update).

Allow duplicates.

Can store different data types (int, string, float, even another list).

üîπ Creating a List
# Empty list
my_list = []

# List with integers
numbers = [1, 2, 3, 4, 5]

# List with mixed data types
mixed = [10, "Hello", 3.14, True]

# Nested list
nested = [1, [2, 3], [4, 5]]

üîπ Accessing List Items

Lists are indexed (starting from 0).

fruits = ["apple", "banana", "cherry"]

print(fruits[0])   # apple (first element)
print(fruits[1])   # banana
print(fruits[-1])  # cherry (last element)

üîπ Changing List Items
fruits = ["apple", "banana", "cherry"]
fruits[1] = "blueberry"   # change banana ‚Üí blueberry
print(fruits)  # ['apple', 'blueberry', 'cherry']

üîπ Adding Items
fruits = ["apple", "banana"]

fruits.append("cherry")      # add at the end
fruits.insert(1, "orange")   # add at index 1
print(fruits)  # ['apple', 'orange', 'banana', 'cherry']

üîπ Removing Items
fruits = ["apple", "banana", "cherry"]

fruits.remove("banana")   # removes 'banana'
print(fruits)  # ['apple', 'cherry']

fruits.pop()   # removes last item
print(fruits)  # ['apple']

fruits.clear()  # removes all items
print(fruits)   # []

üîπ List Operations
nums = [1, 2, 3, 4]

print(len(nums))        # 4 (length of list)
print(nums + [5, 6])    # Concatenation ‚Üí [1, 2, 3, 4, 5, 6]
print(nums * 2)         # Repetition ‚Üí [1, 2, 3, 4, 1, 2, 3, 4]

üîπ Looping Through a List
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

üîπ Useful List Methods
numbers = [3, 1, 4, 1, 5]

numbers.sort()         # sort in ascending order ‚Üí [1, 1, 3, 4, 5]
numbers.reverse()      # reverse order ‚Üí [5, 4, 3, 1, 1]
print(numbers.count(1))  # count occurrences of 1 ‚Üí 2
print(numbers.index(4))  # find index of 4 ‚Üí 1
```

```python
Let‚Äôs go through Python Tuples step by step.

üîπ What is a Tuple?

A tuple is a collection of items, just like a list.

But tuples are immutable, meaning you cannot change, add, or remove elements after creation.

Tuples are defined using parentheses ().

üîπ Creating Tuples
# A tuple of integers
numbers = (1, 2, 3, 4, 5)
print(numbers)

# A tuple with mixed data types
mixed = (1, "hello", 3.14, True)
print(mixed)

# A tuple with nested elements
nested = (1, [2, 3], (4, 5))
print(nested)

# A tuple with one element (must include a comma!)
single = (10,)
print(single)

üîπ Accessing Tuple Items
fruits = ("apple", "banana", "cherry")

print(fruits[0])   # apple
print(fruits[1])   # banana
print(fruits[-1])  # cherry (last element)

üîπ Tuple Slicing
fruits = ("apple", "banana", "cherry", "mango", "grape")

print(fruits[1:3])   # ('banana', 'cherry')
print(fruits[:4])    # ('apple', 'banana', 'cherry', 'mango')
print(fruits[-2:])   # ('mango', 'grape')

üîπ Tuple Methods

Tuples are immutable, so only a few methods exist:

nums = (1, 2, 2, 3, 4, 2)

print(nums.count(2))   # 3 (how many times 2 appears)
print(nums.index(3))   # 3 (first index of 3)

üîπ Tuple Packing and Unpacking
# Packing
person = ("John", 25, "Engineer")

# Unpacking
name, age, profession = person
print(name)       # John
print(age)        # 25
print(profession) # Engineer


‚úÖ Key Points to Remember:

Tuples are like lists, but immutable.

Use () to create tuples.

Can store different data types.

Useful when you want data that should not be changed (like constants, coordinates, database rows).
```

```python
Let‚Äôs go over Python Sets step by step:

üîπ What is a Set in Python?

A set is a collection of unordered, unchangeable (immutable items), and unindexed elements.

Duplicates are not allowed.

Sets are very useful when you want to store unique items.

üîπ Creating a Set
# Creating a set
fruits = {"apple", "banana", "cherry"}
print(fruits)
# Output: {'banana', 'cherry', 'apple'}  (order may vary)

# A set with numbers
numbers = {1, 2, 3, 4, 5}
print(numbers)

# Creating an empty set
empty_set = set()
print(empty_set, type(empty_set))


‚ö†Ô∏è Note: {} creates an empty dictionary, not a set. Use set() instead.

üîπ Properties of Sets

Unordered ‚Üí No guarantee about order.

No duplicates ‚Üí Duplicate values are removed automatically.

Mutable ‚Üí You can add or remove elements.

Example:

sample_set = {1, 2, 2, 3, 4, 4}
print(sample_set)  # Output: {1, 2, 3, 4}

üîπ Adding and Removing Items
fruits = {"apple", "banana"}

# Add one item
fruits.add("cherry")
print(fruits)

# Add multiple items
fruits.update(["orange", "mango"])
print(fruits)

# Remove an item
fruits.remove("banana")   # ‚ùå error if not found
print(fruits)

# Discard an item (no error if missing)
fruits.discard("kiwi")

# Pop removes a random element
removed_item = fruits.pop()
print("Removed:", removed_item)
print(fruits)

# Clear all items
fruits.clear()
print(fruits)   # Output: set()

üîπ Set Operations (like in Math)
A = {1, 2, 3}
B = {3, 4, 5}

# Union (A ‚à™ B)
print(A | B)       # {1, 2, 3, 4, 5}

# Intersection (A ‚à© B)
print(A & B)       # {3}

# Difference (A - B)
print(A - B)       # {1, 2}

# Symmetric Difference (A Œî B ‚Üí items in A or B but not both)
print(A ^ B)       # {1, 2, 4, 5}

üîπ Checking Membership
fruits = {"apple", "banana", "cherry"}
print("apple" in fruits)   # True
print("mango" in fruits)   # False


‚úÖ Use cases of Sets:

Removing duplicates from a list.

Fast membership testing (in is faster in sets than lists).

Mathematical set operations.
```

```python
Let‚Äôs go step by step and understand Python Dictionaries.

üìò Python Dictionaries

A dictionary in Python is used to store key-value pairs.

It is unordered, mutable (can be changed), and does not allow duplicate keys.

Defined using curly braces { }.

‚úÖ Example of a Dictionary
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 22,
    "course": "Data Science",
    "is_graduated": False
}

print(student)


üìå Output:

{'name': 'Alice', 'age': 22, 'course': 'Data Science', 'is_graduated': False}

üîπ Accessing Dictionary Items
print(student["name"])       # Alice
print(student.get("course")) # Data Science

üîπ Changing Values
student["age"] = 23
print(student)  # {'name': 'Alice', 'age': 23, 'course': 'Data Science', 'is_graduated': False}

üîπ Adding Items
student["grade"] = "A"
print(student)

üîπ Removing Items
student.pop("course")   # removes 'course'
del student["age"]      # removes 'age'
print(student)

üîπ Looping Through a Dictionary
# Keys
for key in student:
    print(key)

# Values
for value in student.values():
    print(value)

# Key and Value
for key, value in student.items():
    print(key, ":", value)

üîπ Nested Dictionary (Dictionary inside Dictionary)
school = {
    "student1": {"name": "Alice", "age": 22},
    "student2": {"name": "Bob", "age": 24}
}

print(school["student1"]["name"])  # Alice


‚úÖ Summary:

Dictionaries use keys (must be unique, immutable like strings or numbers).

Values can be of any data type (string, int, list, another dictionary, etc.).

Useful for structured data like JSON.
```

```python
Let‚Äôs go through Python if..else statements step by step.

üìò Python if..else

Used for decision making in Python.

Executes different code blocks depending on conditions.

Conditions are usually boolean expressions (True / False).

‚úÖ Basic if
x = 10

if x > 5:
    print("x is greater than 5")


üìå Output:

x is greater than 5

‚úÖ if..else
x = 3

if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")


üìå Output:

x is not greater than 5

‚úÖ if..elif..else
x = 10

if x > 20:
    print("x is greater than 20")
elif x > 5:
    print("x is greater than 5 but not more than 20")
else:
    print("x is 5 or less")


üìå Output:

x is greater than 5 but not more than 20

‚úÖ Short-Hand if (One-Liner)
x = 5

if x == 5: print("x is 5")

‚úÖ Short-Hand if..else (Ternary Operator)
x = 7

result = "Even" if x % 2 == 0 else "Odd"
print(result)


üìå Output:

Odd

‚úÖ Nested if
x = 18

if x > 10:
    if x < 20:
        print("x is between 10 and 20")


üìå Output:

x is between 10 and 20


‚úÖ Summary:

if ‚Üí checks condition.

elif ‚Üí multiple conditions.

else ‚Üí runs when no condition is true.

Can be nested or written in short form.
```

```python
let‚Äôs learn Python‚Äôs match statement (structural pattern matching, Python ‚â• 3.10) step by step. Think of it as a super-powered switch, plus smart destructuring.

0) Quick rules

One subject is evaluated once: match subject: ...

First matching case wins (no fall-through).

Use _ as a ‚Äúcatch-all‚Äù default.

Bare names in a pattern capture values; to compare to a constant, use a dotted name (e.g., Color.RED) or a literal.

1) Literals + default
lang = "fr"

match lang:
    case "en":
        print("Hello")
    case "fr":
        print("Bonjour")
    case _:
        print("Hi (default)")

2) Multiple alternatives (|)
code = 404
match code:
    case 200 | 201 | 204:
        print("Success-ish")
    case 400 | 404:
        print("Client error")
    case _:
        print("Other")

3) Guards (if after a pattern)
age = 17
match age:
    case a if a < 13:
        print("Child")
    case a if 13 <= a < 18:
        print("Teen")
    case _:
        print("Adult")

4) Variable capture (bind parts of the subject)
val = ("USD", 99.5)

match val:
    case (currency, amount):
        print(f"{amount} in {currency}")   # binds two names


‚ö†Ô∏è case name: captures the value; it does not compare to an existing variable named name.

5) Sequence patterns (lists/tuples) + ‚Äúrest‚Äù (*)
row = ["Alice", 92, 88, 95]

match row:
    case [name, *scores] if scores:
        print(name, "avg:", sum(scores)/len(scores))
    case []:
        print("Empty row")

6) Mapping (dict) patterns
record = {"type": "user", "name": "Sharmila", "active": True}

match record:
    case {"type": "user", "name": n, "active": True}:
        print(f"Active user: {n}")
    case {"type": "user", "name": n}:
        print(f"User: {n} (status unknown)")
    case {"type": t, **rest}:
        print("Other type:", t, rest)

7) Class patterns (dataclasses / namedtuples / classes with __match_args__)
from dataclasses import dataclass

@dataclass
class Point:
    x: int
    y: int

p = Point(3, 3)

match p:
    case Point(0, 0):
        print("Origin")
    case Point(x, y) if x == y:
        print("On y = x:", x)
    case Point(x=3, y=yval):
        print("x is 3; y =", yval)


Dataclasses and namedtuples work out of the box. For custom classes, you can define __match_args__ to control positional matching.

8) OR-patterns + grouping
token = ("CMD", "EXIT")

match token:
    case ("CMD", "HELP") | ("CMD", "H"):
        print("Show help")
    case ("CMD", "EXIT"):
        print("Goodbye")
    case _:
        print("Unknown command")

9) ‚Äúas‚Äù capture (grab the whole matched value)
item = ["err", 500, "timeout"]

match item:
    case ["err", code, msg] as whole if code >= 500:
        print("Server error:", whole)   # keeps the full list

10) Practical mini example: routing a JSON-like payload
msg = {"event": "purchase", "user": "sc", "amount": 49.0}

match msg:
    case {"event": "login", "user": u}:
        print(f"{u} logged in")
    case {"event": "purchase", "user": u, "amount": a} if a > 0:
        print(f"{u} purchased ${a}")
    case {"event": "purchase"}:
        print("Malformed purchase")
    case _:
        print("Unknown event")

Cheat sheet (when to use what)

Exact value ‚Üí case "OK":

One of many ‚Üí case 400 | 404:

Default ‚Üí case _:

Destructure tuple/list ‚Üí case [head, *rest]:

Match dict shape ‚Üí case {"k": v, **rest}:

Match class ‚Üí case Point(x, y):

Extra condition ‚Üí case p if cond(p):

Capture whole ‚Üí case pattern as name:
```

```python
Let‚Äôs go step by step through Python while loops.

üîπ What is a while loop?

A while loop in Python keeps executing a block of code as long as the condition is True.

Syntax:
while condition:
    # code to run


condition ‚Üí evaluated before each iteration.

If the condition is True, the loop runs.

If the condition is False, the loop stops.

üîπ Example 1: Basic while loop
count = 1

while count <= 5:
    print("Number:", count)
    count += 1


‚úÖ Output:

Number: 1
Number: 2
Number: 3
Number: 4
Number: 5


üëâ The loop runs until count becomes 6.

üîπ Example 2: while loop with break

break stops the loop immediately.

num = 1

while num <= 10:
    if num == 5:
        print("Breaking at:", num)
        break
    print("Number:", num)
    num += 1


‚úÖ Output:

Number: 1
Number: 2
Number: 3
Number: 4
Breaking at: 5

üîπ Example 3: while loop with continue

continue skips the rest of the code in the loop and goes to the next iteration.

num = 0

while num < 6:
    num += 1
    if num == 3:
        continue
    print("Number:", num)


‚úÖ Output:

Number: 1
Number: 2
Number: 4
Number: 5
Number: 6


üëâ 3 is skipped.

üîπ Example 4: Real-life example ‚Äì Library Book Borrowing
books_left = 3

while books_left > 0:
    print("You borrowed a book! üìö")
    books_left -= 1

print("No books left!")


‚úÖ Output:

You borrowed a book! üìö
You borrowed a book! üìö
You borrowed a book! üìö
No books left!


‚ö†Ô∏è Important Rule:
Always make sure the while loop condition will eventually become False (otherwise you‚Äôll create an infinite loop).
```

```python
Let‚Äôs go step by step with Python for loops.

üîπ What is a for loop?

A for loop is used to iterate over a sequence (like a list, tuple, string, dictionary, or range) element by element.

Unlike while, it iterates over items, not just based on a condition.

Syntax:
for variable in sequence:
    # code to run

üîπ Example 1: Iterating over a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)


‚úÖ Output:

apple
banana
cherry

üîπ Example 2: Iterating over a string
for letter in "Python":
    print(letter)


‚úÖ Output:

P
y
t
h
o
n

üîπ Example 3: Using range()

range() generates a sequence of numbers.

# 0 to 4
for i in range(5):
    print(i)

# 2 to 6
for i in range(2, 7):
    print(i)

# 1 to 10 with step 2
for i in range(1, 11, 2):
    print(i)

üîπ Example 4: Iterating over a dictionary
student = {"name": "Alice", "age": 22, "course": "Math"}

# keys
for key in student:
    print(key)

# values
for value in student.values():
    print(value)

# key-value pairs
for key, value in student.items():
    print(key, ":", value)

üîπ Example 5: Using break and continue
for num in range(1, 6):
    if num == 3:
        continue   # skips 3
    if num == 5:
        break      # stops loop at 5
    print(num)


‚úÖ Output:

1
2
4

üîπ Example 6: Real-life ‚Äì Library Books
books = ["Math", "Science", "English", "History"]

for book in books:
    print(f"You borrowed the {book} book üìö")


‚úÖ Output:

You borrowed the Math book üìö
You borrowed the Science book üìö
You borrowed the English book üìö
You borrowed the History book üìö


‚úÖ Key Points:

for is great for sequences (lists, strings, tuples, dicts, ranges).

Can use break to exit early, continue to skip.

Works well for real-life tasks like processing lists of items.
```

```python
Let‚Äôs go step by step and learn about Python functions.
# functions are very very important in real time applications

üîπ What is a Function in Python?

A function is a block of reusable code that performs a specific task.

Functions help to avoid repeating code and make programs organized and readable.

Syntax:
def function_name(parameters):
    """
    Optional docstring explaining the function
    """
    # Code block
    return result  # optional

üîπ Example 1: Simple Function
def greet():
    print("Hello! Welcome to Python.")

# Call the function
greet()


‚úÖ Output:

Hello! Welcome to Python.

üîπ Example 2: Function with Parameters
def greet_person(name):
    print(f"Hello, {name}! Welcome to Python.")

greet_person("Alice")
greet_person("Bob")


‚úÖ Output:

Hello, Alice! Welcome to Python.
Hello, Bob! Welcome to Python.

üîπ Example 3: Function with Return Value
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print("Sum:", result)


‚úÖ Output:

Sum: 8

üîπ Example 4: Default Parameters
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet("Alice")  # Hello, Alice!
greet()         # Hello, Guest!

üîπ Example 5: Keyword Arguments
def describe_person(name, age, country):
    print(f"{name} is {age} years old and lives in {country}.")

describe_person(name="Alice", age=25, country="India")
describe_person(age=30, country="USA", name="Bob")

üîπ Example 6: Arbitrary Arguments (*args, **kwargs)
# *args ‚Üí arbitrary number of positional arguments
def add_numbers(*args):
    return sum(args)

print(add_numbers(1, 2, 3))        # 6
print(add_numbers(5, 10, 15, 20))  # 50

# **kwargs ‚Üí arbitrary number of keyword arguments
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, country="India")

üîπ Example 7: Real-Life Function ‚Äì Loan Eligibility
def check_loan_eligibility(name, loan_amount, credit_score):
    MIN_CREDIT_SCORE = 600
    MAX_LOAN_AMOUNT = 50000

    if credit_score >= MIN_CREDIT_SCORE and loan_amount <= MAX_LOAN_AMOUNT:
        return f"Congratulations {name}, your loan is APPROVED ‚úÖ"
    elif credit_score < MIN_CREDIT_SCORE:
        return f"Sorry {name}, your credit score is too low. Loan REJECTED ‚ùå"
    else:
        return f"Sorry {name}, requested loan amount exceeds limit. Loan REJECTED ‚ùå"

print(check_loan_eligibility("Alice", 30000, 720))
print(check_loan_eligibility("Bob", 60000, 680))


‚úÖ Key Points:

Functions help organize code and reduce repetition.

Parameters can be required, default, or arbitrary.

Functions can return values or just perform actions.

Always give meaningful names to functions and parameters.
```

```python
# Mini College Student Application & Attendance Program üéì

# Dictionary to store student info
students = {}

# Function to add a student
def add_student():
    name = input("Enter student name: ").title()
    roll_no = input("Enter roll number: ").upper()
    students[roll_no] = {"name": name, "attendance": 0}
    print(f"Student {name} (Roll No: {roll_no}) added successfully!\n")

# Function to mark attendance
def mark_attendance():
    roll_no = input("Enter roll number to mark attendance: ").upper()
    if roll_no in students:
        students[roll_no]["attendance"] += 1
        print(f"Attendance marked for {students[roll_no]['name']}. Total: {students[roll_no]['attendance']}\n")
    else:
        print("Student not found!\n")

# Function to view all students
def view_students():
    if not students:
        print("No students added yet.\n")
        return
    print("\n--- Student List & Attendance ---")
    for roll_no, info in students.items():
        print(f"Roll No: {roll_no} | Name: {info['name']} | Attendance: {info['attendance']}")
    print()

# Main program loop
def main():
    while True:
        print("===== College Student Management =====")
        print("1. Add Student")
        print("2. Mark Attendance")
        print("3. View Students")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            add_student()
        elif choice == "2":
            mark_attendance()
        elif choice == "3":
            view_students()
        elif choice == "4":
            print("Exiting program. Goodbye! üéì")
            break
        else:
            print("Invalid choice! Please enter 1-4.\n")

# Run the program
main()
```

```python
#Let‚Äôs create a mini interactive health application to check for diabetes risk using Python functions üçéüíâ.

#This program will:

#Ask the user for basic information (age, weight, family history, symptoms).

#Evaluate diabetes risk based on simple criteria.

#Provide recommendations.

# Mini Health Application: Diabetes Risk Checker üçéüíâ

# Function to get user details
def get_user_info():
    name = input("Enter your name: ").title()
    age = int(input("Enter your age: "))
    weight = float(input("Enter your weight (kg): "))
    family_history = input("Do you have a family history of diabetes? (yes/no): ").lower()
    frequent_thirst = input("Do you feel frequent thirst? (yes/no): ").lower()
    frequent_urination = input("Do you have frequent urination? (yes/no): ").lower()
    fatigue = input("Do you feel fatigue often? (yes/no): ").lower()

    return {
        "name": name,
        "age": age,
        "weight": weight,
        "family_history": family_history,
        "frequent_thirst": frequent_thirst,
        "frequent_urination": frequent_urination,
        "fatigue": fatigue
    }

# Function to evaluate diabetes risk
def evaluate_risk(user):
    risk_score = 0

    # Age factor
    if user["age"] >= 45:
        risk_score += 1

    # Weight factor (simple BMI-based approximation, assuming height ~1.7m)
    bmi = user["weight"] / (1.7 ** 2)
    if bmi >= 25:
        risk_score += 1

    # Family history
    if user["family_history"] == "yes":
        risk_score += 1

    # Symptoms
    if user["frequent_thirst"] == "yes":
        risk_score += 1
    if user["frequent_urination"] == "yes":
        risk_score += 1
    if user["fatigue"] == "yes":
        risk_score += 1

    # Evaluate risk
    if risk_score >= 5:
        risk_level = "High Risk"
        advice = "Consult a doctor immediately and get tested."
    elif risk_score >= 3:
        risk_level = "Moderate Risk"
        advice = "Monitor lifestyle, diet, and consider checking with a doctor."
    else:
        risk_level = "Low Risk"
        advice = "Maintain healthy lifestyle and regular checkups."

    return risk_level, advice

# Function to display result
def display_result(user, risk_level, advice):
    print("\n===== Diabetes Risk Report =====")
    print(f"Name: {user['name']}")
    print(f"Age: {user['age']}")
    print(f"Weight: {user['weight']} kg")
    print(f"Risk Level: {risk_level}")
    print(f"Advice: {advice}")
    print("================================\n")

# Main program loop
def main():
    print("Welcome to the Diabetes Risk Checker üçéüíâ")
    while True:
        user = get_user_info()
        risk_level, advice = evaluate_risk(user)
        display_result(user, risk_level, advice)

        another = input("Do you want to check another person? (yes/no): ").lower()
        if another != "yes":
            print("Thank you for using the Diabetes Risk Checker. Stay healthy! üíö")
            break

# Run the program
main()

```

```python
Let‚Äôs go step by step and learn about Python lambda functions.

üîπ What is a Lambda Function?

A lambda function is a small anonymous function in Python.

It can have any number of arguments, but only one expression.

The expression is automatically returned.

Syntax:

lambda arguments: expression

üîπ Example 1: Basic Lambda
# Regular function
def square(x):
    return x ** 2

print(square(5))  # Output: 25

# Lambda function
square_lambda = lambda x: x ** 2
print(square_lambda(5))  # Output: 25

üîπ Example 2: Lambda with Multiple Arguments
add = lambda a, b: a + b
print(add(3, 5))  # Output: 8

multiply = lambda a, b, c: a * b * c
print(multiply(2, 3, 4))  # Output: 24

üîπ Example 3: Lambda in a List
numbers = [1, 2, 3, 4, 5]

# Square all numbers using map + lambda
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Output: [1, 4, 9, 16, 25]

üîπ Example 4: Lambda with filter
numbers = [1, 2, 3, 4, 5, 6]

# Filter even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]

üîπ Example 5: Lambda with sorted
students = [("Alice", 22), ("Bob", 20), ("Charlie", 25)]

# Sort by age
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
# Output: [('Bob', 20), ('Alice', 22), ('Charlie', 25)]

üîπ Example 6: Real-Life Mini Example ‚Äì Loan Approval Score
# Simple score calculator using lambda
calculate_score = lambda credit_score, income: credit_score * 0.7 + income * 0.3

score = calculate_score(720, 5000)
print("Loan approval score:", score)


‚úÖ Key Points:

lambda functions are compact and good for short operations.

Often used with map, filter, and sorted.

Avoid using lambda for complex logic‚Äîregular functions are better then.
```

```python
Let‚Äôs go step by step on Python arrays.

üîπ 1) What is an Array in Python?

An array is a collection of items of the same data type stored at contiguous memory locations.

Python lists can act like arrays but are more flexible (can hold different types).

For true arrays, Python provides the array module.

Arrays are useful when you need efficient storage and operations on large datasets of the same type.

üîπ 2) Using the array Module
import array

# Create an array of integers
numbers = array.array('i', [1, 2, 3, 4, 5])

print(numbers)         # Output: array('i', [1, 2, 3, 4, 5])
print(numbers[0])      # Access first element: 1
print(numbers[-1])     # Access last element: 5


Notes:

'i' ‚Üí type code for signed integer.

Other common type codes:

'f' ‚Üí float

'd' ‚Üí double

'u' ‚Üí Unicode character

üîπ 3) Adding Elements to Array
numbers.append(6)   # Add at the end
numbers.insert(2, 99)  # Insert 99 at index 2
print(numbers)      # Output: array('i', [1, 2, 99, 3, 4, 5, 6])

üîπ 4) Removing Elements from Array
numbers.remove(99)    # Remove value 99
numbers.pop()         # Remove last element
print(numbers)        # Output: array('i', [1, 2, 3, 4, 5])

üîπ 5) Iterating Over an Array
for num in numbers:
    print(num)


‚úÖ Output:

1
2
3
4
5

üîπ 6) Array Operations
# Sum
print("Sum:", sum(numbers))    # 15

# Length
print("Length:", len(numbers))  # 5

# Index
print("Index of 3:", numbers.index(3))  # 2

üîπ 7) Real-Life Example ‚Äì Student Marks Array
from array import array

# Marks of students
marks = array('i', [75, 88, 92, 60, 85])

# Average marks
average = sum(marks) / len(marks)
print("Average Marks:", average)

# Find students who scored above 80
high_scores = [mark for mark in marks if mark > 80]
print("High Scores:", high_scores)


‚úÖ Output:

Average Marks: 80.0
High Scores: [88, 92, 85]

‚ö†Ô∏è Notes:

If you need more advanced operations, use NumPy arrays (numpy.array), which are faster and support mathematical operations on arrays efficiently.
```

```python
Let‚Äôs go step by step into Python OOP (Object-Oriented Programming).

üîπ What is OOP?

Object-Oriented Programming (OOP) is a way of structuring programs so that properties and behaviors are bundled into individual objects.

In Python, everything is an object (numbers, strings, functions, etc.).

OOP helps write clean, reusable, and modular code.

üîπ Core OOP Concepts in Python:

Class - A blueprint for creating objects.

Object (Instance)- A specific instance created from a class.

Attributes - Variables that belong to a class/object.

Methods -Functions defined inside a class.

Inheritance-Allows one class to inherit features from another.

Encapsulation-Restricting direct access to class attributes.

Polymorphism- Same function/method behaves differently depending on the object.

üîπ Simple Example in Python:
# Defining a Class
class Student:
    # Constructor (initializer method)
    def __init__(self, name, marks):
        self.name = name      # attribute
        self.marks = marks    # attribute

    # Method to display details
    def display(self):
        print(f"Student: {self.name}, Marks: {self.marks}")

    # Method to check pass/fail
    def result(self):
        if self.marks >= 40:
            print(f"{self.name} has Passed ‚úÖ")
        else:
            print(f"{self.name} has Failed ‚ùå")

# Creating Objects
s1 = Student("Alice", 85)
s2 = Student("Bob", 32)

# Using methods
s1.display()
s1.result()

s2.display()
s2.result()

üñ• Example Output:
Student: Alice, Marks: 85
Alice has Passed ‚úÖ
Student: Bob, Marks: 32
Bob has Failed ‚ùå

üîπ Inheritance Example:
# Parent Class
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, my name is {self.name}")

# Child Class (inherits from Person)
class Student(Person):
    def __init__(self, name, marks):
        super().__init__(name)  # calling parent constructor
        self.marks = marks

    def show_marks(self):
        print(f"{self.name} scored {self.marks} marks")

# Create object
s = Student("Charlie", 92)
s.greet()        # Inherited method
s.show_marks()   # Child method


‚úÖ With OOP, you can build scalable applications like banking systems, health apps, school management systems, etc.
```

```python
here‚Äôs a step-by-step tour of Python OOP with tiny, runnable examples for each core concept. You can paste each block into a Python file or notebook and run them one by one.

1) Class ‚Äî the blueprint

A class groups related data and behavior.

# Step 1: Class (blueprint)
class Student:
    # Constructor initializes new objects
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

2) Object (Instance) ‚Äî a thing made from the class

Create instances of the class.

# Step 2: Object / Instance
s1 = Student("Alice", 85)
s2 = Student("Bob", 72)

print(type(s1))   # <class '__main__.Student'>
print(s1.name)    # Alice
print(s2.marks)   # 72

3) Attributes ‚Äî data stored on the object

Attributes are just variables on the instance (or class).

# Step 3: Attributes
print(s1.name, s1.marks)  # Alice 85
s1.marks = 90             # update an attribute
print(s1.marks)           # 90

4) Methods ‚Äî functions defined inside a class

Methods operate on the object (they receive self).

# Step 4: Methods
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def greet(self):                     # instance method
        return f"Hi, I'm {self.name}!"

    def is_pass(self):                   # simple behavior
        return self.marks >= 40

s = Student("Carla", 55)
print(s.greet())     # Hi, I'm Carla!
print(s.is_pass())   # True

5) Inheritance ‚Äî reuse & specialize behavior

A child class inherits from a parent class and can add/override.

# Step 5: Inheritance
class Person:
    def __init__(self, name):
        self.name = name

    def intro(self):
        return f"I'm {self.name}"

class Student(Person):                    # Student inherits Person
    def __init__(self, name, marks):
        super().__init__(name)            # call parent constructor
        self.marks = marks

    def intro(self):                      # override (specialize)
        base = super().intro()
        return f"{base}, and I scored {self.marks}"

p = Person("Dana")
st = Student("Evan", 91)
print(p.intro())   # I'm Dana
print(st.intro())  # I'm Evan, and I scored 91

6) Encapsulation ‚Äî control access to internal state

Use naming conventions and properties to hide and validate data.

# Step 6: Encapsulation
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance    # double underscore => name-mangled (pseudo-private)

    @property
    def balance(self):              # read-only view (or add checks in setter)
        return self.__balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit must be positive")
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal must be positive")
        if amount > self.__balance:
            raise ValueError("Insufficient funds")
        self.__balance -= amount

acct = BankAccount("Grace", 100)
acct.deposit(50)
print(acct.balance)     # 150

# Direct access blocked by name mangling:
# print(acct.__balance)  # AttributeError

# (Not recommended) You *could* reach it via _BankAccount__balance, but that breaks encapsulation.


Why this matters: callers use deposit/withdraw instead of mutating the balance directly, so you can enforce rules safely.

7) Polymorphism ‚Äî same interface, different behavior

Two flavors:

Overriding (via inheritance).

Duck typing (any object with the needed method works).

# Step 7: Polymorphism

class Animal:
    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

def announce(animal):               # works with any object that has .speak()
    # duck typing: only behavior matters
    print(animal.speak())

animals = [Dog(), Cat(), Animal()]
for a in animals:
    announce(a)
# Output:
# Woof!
# Meow!
# ...

TL;DR

Class: blueprint

Object: built from a class

Attributes: data on the object

Methods: functions on the object

Inheritance: reuse/extend behavior

Encapsulation: protect internal state

Polymorphism: one interface, many implementations
```

```python
Let‚Äôs walk through Inheritance in Python step by step with a simple example.

üîπ What is Inheritance?

Inheritance lets a class (child) reuse attributes and methods from another class (parent).

Parent/Base Class ‚Üí General blueprint.

Child/Derived Class ‚Üí Extends/overrides features of parent.

üîπ Example: Vehicles üöóüöå
# Parent class (Base)
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def drive(self):
        print(f"{self.brand} is moving...")

# Child class (inherits from Vehicle)
class Car(Vehicle):
    def __init__(self, brand, model):
        # Call parent constructor
        super().__init__(brand)
        self.model = model

    # Override method
    def drive(self):
        print(f"{self.brand} {self.model} is driving on the road!")

# Another Child class
class Bus(Vehicle):
    def drive(self):
        print(f"{self.brand} bus is carrying passengers!")

# Demo
def main():
    v = Vehicle("Generic Vehicle")
    v.drive()

    c = Car("Toyota", "Camry")
    c.drive()

    b = Bus("Volvo")
    b.drive()

if __name__ == "__main__":
    main()

üîπ Output:
Generic Vehicle is moving...
Toyota Camry is driving on the road!
Volvo bus is carrying passengers!

‚úÖ Key Points

Car and Bus inherit from Vehicle.

super().__init__(brand) calls the parent constructor so we don‚Äôt rewrite it.

Each child can override methods (drive) to behave differently.
```

```python
Let‚Äôs explore Python Iterators step by step.

üîπ What is an Iterator?

An iterator is an object that lets you loop through elements one at a time.
It must implement:

__iter__() ‚Üí returns the iterator object.

__next__() ‚Üí returns the next value (and raises StopIteration when no more items).

üîπ Example: Simple Iterator
class CountUpTo:
    def __init__(self, max_value):
        self.max_value = max_value
        self.current = 1

    def __iter__(self):
        return self   # the iterator is self

    def __next__(self):
        if self.current <= self.max_value:
            value = self.current
            self.current += 1
            return value
        else:
            raise StopIteration   # no more elements

üîπ Using the Iterator
# Create iterator object
counter = CountUpTo(5)

# Manual iteration
print(next(counter))  # 1
print(next(counter))  # 2

# Loop automatically calls __next__()
for num in CountUpTo(5):
    print(num)

üîπ Output
1
2
1
2
3
4
5

üîπ Tiny Interactive Program üéÆ

Here‚Äôs a menu-driven version so you can play with an iterator:

class CountUpTo:
    def __init__(self, max_value):
        self.max_value = max_value
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= self.max_value:
            value = self.current
            self.current += 1
            return value
        else:
            raise StopIteration

def main():
    print("=== Iterator Demo ===")
    n = int(input("Enter a max number to count up to: "))
    counter = CountUpTo(n)

    while True:
        choice = input("\nOptions:\n1. Get next number\n2. Reset iterator\n3. Quit\nChoose: ")

        if choice == "1":
            try:
                print("Next:", next(counter))
            except StopIteration:
                print("Iterator is exhausted! Reset to start again.")
        elif choice == "2":
            counter = CountUpTo(n)
            print("Iterator reset.")
        elif choice == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice, try again!")

if __name__ == "__main__":
    main()
```

```python
Let‚Äôs break down Python Polymorphism step by step.

üîπ What is Polymorphism?

The word polymorphism means "many forms".
In Python OOP, it refers to the ability to use the same method name for different types of objects, and each object responds in its own way.

üîπ Example 1: Polymorphism with Functions
def add(x, y, z=0):
    return x + y + z

print(add(2, 3))        # 5
print(add(2, 3, 4))     # 9
print(add("Hello", " World"))  # Hello World


üëâ Same function add() works differently depending on the type and number of arguments.

üîπ Example 2: Polymorphism with Classes
class Dog:
    def sound(self):
        return "Woof!"

class Cat:
    def sound(self):
        return "Meow!"

# Polymorphism in action
for animal in (Dog(), Cat()):
    print(animal.sound())

‚úÖ Output
Woof!
Meow!


üëâ Both Dog and Cat classes have a sound() method, but the output depends on the object.

üîπ Example 3: Polymorphism with Inheritance (Method Overriding)
class Vehicle:
    def move(self):
        return "This vehicle moves"

class Car(Vehicle):
    def move(self):
        return "Car drives on road"

class Boat(Vehicle):
    def move(self):
        return "Boat sails on water"

# Same method name "move" behaves differently
for v in (Vehicle(), Car(), Boat()):
    print(v.move())

‚úÖ Output
This vehicle moves
Car drives on road
Boat sails on water

üîπ Tiny Interactive Program üéÆ

Here‚Äôs a menu-driven example showing polymorphism in action:

class Bird:
    def speak(self):
        return "Chirp!"

class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

def main():
    animals = {"1": Bird(), "2": Dog(), "3": Cat()}

    while True:
        print("\nChoose an animal to hear its sound:")
        print("1. Bird\n2. Dog\n3. Cat\n4. Quit")
        choice = input("Enter choice: ")

        if choice in animals:
            print("Sound:", animals[choice].speak())
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```

```python
Let‚Äôs go step by step on Python Scope.

üîπ What is Scope?

Scope determines where a variable can be accessed in Python.

There are four main types of scope (often remembered as LEGB rule):

Local (L) ‚Äì Inside a function or block.

Enclosing (E) ‚Äì In enclosing function (nested functions).

Global (G) ‚Äì At the top-level of the file or module.

Built-in (B) ‚Äì Names pre-defined by Python (like print(), len()).

üîπ 1) Local Scope

Variables defined inside a function are local to that function.

def my_function():
    x = 10  # local variable
    print("Inside function:", x)

my_function()
# print(x)  # ‚ùå Error: x is not accessible outside the function

üîπ 2) Global Scope

Variables defined outside functions are global and can be accessed anywhere in the file.

x = 50  # global variable

def show():
    print("Inside function:", x)

show()
print("Outside function:", x)


‚úÖ Output:

Inside function: 50
Outside function: 50

üîπ 3) Enclosing Scope (Nested Functions)

Variables in an outer function can be accessed by inner functions.

def outer():
    y = 20  # enclosing variable

    def inner():
        print("Inner function sees y =", y)

    inner()

outer()

üîπ 4) Global Keyword

To modify a global variable inside a function, use the global keyword.

count = 0

def increment():
    global count
    count += 1
    print("Inside function:", count)

increment()
print("Outside function:", count)

üîπ 5) Nonlocal Keyword

To modify an enclosing variable in nested functions, use nonlocal.

def outer():
    num = 5

    def inner():
        nonlocal num
        num += 10
        print("Inner num:", num)

    inner()
    print("Outer num:", num)

outer()


‚úÖ Output:

Inner num: 15
Outer num: 15
```

```python
Let‚Äôs go step by step on Python JSON.

üîπ What is JSON?

JSON (JavaScript Object Notation) is a lightweight data format used to store and exchange data.
Python can work with JSON using the built-in json module.

1) Import the module
import json

2) Python to JSON (Serialization)

Convert Python objects to JSON string using json.dumps().

import json

data = {
    "name": "Alice",
    "age": 25,
    "city": "New York",
    "skills": ["Python", "SQL", "ML"]
}

json_string = json.dumps(data)
print(json_string)


‚úÖ Output:

{"name": "Alice", "age": 25, "city": "New York", "skills": ["Python", "SQL", "ML"]}

3) JSON to Python (Deserialization)

Convert JSON string back to Python object using json.loads().

python_data = json.loads(json_string)
print(python_data)
print("Name:", python_data["name"])
print("Skills:", python_data["skills"])

4) Working with JSON Files
Writing to a file
with open("data.json", "w") as f:
    json.dump(data, f)

Reading from a file
with open("data.json", "r") as f:
    data_from_file = json.load(f)
    print(data_from_file)

5) Tiny Interactive Program: JSON Manager
import json

def main():
    data = {}

    while True:
        print("\n=== JSON Manager ===")
        print("1. Add entry")
        print("2. Show JSON")
        print("3. Save to file")
        print("4. Load from file")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            key = input("Enter key: ")
            value = input("Enter value: ")
            data[key] = value
            print("‚úÖ Entry added.")

        elif choice == "2":
            print("Current JSON data:")
            print(json.dumps(data, indent=4))

        elif choice == "3":
            with open("data.json", "w") as f:
                json.dump(data, f, indent=4)
            print("‚úÖ Data saved to data.json")

        elif choice == "4":
            with open("data.json", "r") as f:
                data = json.load(f)
            print("‚úÖ Data loaded from data.json")

        elif choice == "5":
            print("Exiting...")
            break

        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()


This program allows the user to:

Add key-value entries

View the JSON structure

Save to a file

Load from a file
```

```python
Let‚Äôs go step by step on Python Regular Expressions (RegEx).

üîπ What is RegEx?

RegEx (Regular Expression) is a powerful tool to search, match, and manipulate strings based on patterns.
Python provides the built-in re module for working with regex.

1) Import the Module
import re

2) Basic Functions
Function	Description
re.match()	Checks for a match at the beginning of the string
re.search()	Searches for a pattern anywhere in the string
re.findall()	Returns all matches as a list
re.sub()	Replaces occurrences of a pattern with a string
3) Example: Matching a Pattern
import re

text = "My email is alice@example.com"
pattern = r"\w+@\w+\.\w+"  # basic email pattern

match = re.search(pattern, text)
if match:
    print("Found email:", match.group())
else:
    print("No match found")


‚úÖ Output:

Found email: alice@example.com

4) Example: Finding All Numbers
text = "I have 2 cats and 3 dogs"
numbers = re.findall(r"\d+", text)
print("Numbers found:", numbers)


‚úÖ Output:

Numbers found: ['2', '3']

5) Example: Replacing Text
text = "Hello 123, hello 456"
new_text = re.sub(r"\d+", "###", text)
print(new_text)


‚úÖ Output:

Hello ###, hello ###

6) Tiny Interactive Program: RegEx Tester
import re

def main():
    print("=== Python RegEx Tester ===")

    while True:
        print("\nOptions:")
        print("1. Search pattern")
        print("2. Find all matches")
        print("3. Replace pattern")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            text = input("Enter text: ")
            pattern = input("Enter pattern: ")
            match = re.search(pattern, text)
            if match:
                print("Match found:", match.group())
            else:
                print("No match found.")

        elif choice == "2":
            text = input("Enter text: ")
            pattern = input("Enter pattern: ")
            matches = re.findall(pattern, text)
            print("Matches:", matches)

        elif choice == "3":
            text = input("Enter text: ")
            pattern = input("Enter pattern: ")
            replacement = input("Enter replacement text: ")
            new_text = re.sub(pattern, replacement, text)
            print("Result:", new_text)

        elif choice == "4":
            print("Exiting...")
            break

        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()


This program allows the user to:

Search a pattern in text

Find all occurrences of a pattern

Replace patterns dynamically
```

```python
Let‚Äôs go step by step on Python try..except for error handling.

üîπ What is try..except?

In Python, errors (exceptions) can occur during program execution.
try..except is used to catch exceptions and handle them gracefully instead of crashing the program.

1) Basic Syntax
try:
    # Code that may raise an exception
    x = int(input("Enter a number: "))
    print("You entered:", x)
except ValueError:
    # Code to handle the exception
    print("Oops! That was not a valid number.")


‚úÖ Example Output:

Enter a number: abc
Oops! That was not a valid number.

2) Catching Multiple Exceptions
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
except ValueError:
    print("Invalid input! Please enter numbers only.")
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print("Result:", result)

3) Using finally

finally block always executes, even if an exception occurs.

try:
    file = open("example.txt", "r")
    content = file.read()
except FileNotFoundError:
    print("File not found!")
finally:
    print("This runs no matter what.")
    # file.close() would normally go here if opened successfully

4) Tiny Interactive Program: Exception Handler
def exception_demo():
    while True:
        print("\n=== Exception Demo ===")
        print("1. Divide numbers")
        print("2. Convert input to integer")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            try:
                a = float(input("Enter numerator: "))
                b = float(input("Enter denominator: "))
                print("Result:", a / b)
            except ZeroDivisionError:
                print("Cannot divide by zero!")
            except ValueError:
                print("Please enter valid numbers.")

        elif choice == "2":
            try:
                x = int(input("Enter an integer: "))
                print("You entered:", x)
            except ValueError:
                print("That's not a valid integer!")

        elif choice == "3":
            print("Exiting...")
            break

        else:
            print("Invalid choice!")

if __name__ == "__main__":
    exception_demo()

üîπ Summary

try ‚Üí Run the code that may raise an exception.

except ‚Üí Catch and handle specific exceptions.

else ‚Üí Runs if no exception occurs.

finally ‚Üí Always runs regardless of exceptions.
```

```python
Let‚Äôs go step by step on Python String Formatting.

üîπ What is String Formatting?

String formatting allows you to insert variables into strings in a readable and flexible way.
Python provides multiple methods for formatting strings.

1) Using % Operator (Old Style)
name = "Alice"
age = 25

print("Hello, %s! You are %d years old." % (name, age))


‚úÖ Output:

Hello, Alice! You are 25 years old.

2) Using str.format() Method
name = "Bob"
age = 30

print("Hello, {}! You are {} years old.".format(name, age))
print("Hello, {1}! You are {0} years old.".format(age, name))  # positional


‚úÖ Output:

Hello, Bob! You are 30 years old.
Hello, Bob! You are 30 years old.

3) Using Named Placeholders
print("Hello, {name}! You are {age} years old.".format(name="Charlie", age=35))


‚úÖ Output:

Hello, Charlie! You are 35 years old.

4) Using f-Strings (Python 3.6+)
name = "David"
age = 28

print(f"Hello, {name}! You are {age} years old.")


‚úÖ Output:

Hello, David! You are 28 years old.


Advantages of f-strings:

Easier to read and write

Supports expressions directly

a = 5
b = 10
print(f"Sum of {a} + {b} = {a + b}")

5) Formatting Numbers
pi = 3.1415926535

# 2 decimal places
print(f"Pi rounded: {pi:.2f}")

# Padding numbers
print(f"{'Name':<10} {'Age':>5}")
print(f"{'Alice':<10} {25:>5}")


‚úÖ Output:

Pi rounded: 3.14
Name         Age
Alice         25

6) Tiny Interactive Program: String Formatter
def string_formatter():
    print("=== Python String Formatter ===")
    while True:
        print("\nOptions:")
        print("1. f-string")
        print("2. str.format()")
        print("3. % operator")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice in ["1", "2", "3"]:
            name = input("Enter your name: ")
            age = input("Enter your age: ")

            if choice == "1":
                print(f"Hello, {name}! You are {age} years old.")
            elif choice == "2":
                print("Hello, {}! You are {} years old.".format(name, age))
            elif choice == "3":
                print("Hello, %s! You are %s years old." % (name, age))

        elif choice == "4":
            print("Exiting...")
            break

        else:
            print("Invalid choice!")

if __name__ == "__main__":
    string_formatter()


This program lets the user practice all three string formatting styles dynamically.
```

```python
Let‚Äôs upgrade the big OOP & Python Concepts menu and integrate the String Formatting demo, so you can interactively explore all concepts from one program.

Here‚Äôs the updated runnable script:

# ==========================
# Python OOP & Concepts Menu
# ==========================

import re
import math
import json
from datetime import date, timedelta

# -----------------------------
# 1. Class & Object Example
# -----------------------------
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show_info(self):
        return f"Car: {self.brand} {self.model}"


# -----------------------------
# 2. Inheritance Example
# -----------------------------
class ElectricCar(Car):
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity

    def show_info(self):
        return f"Electric Car: {self.brand} {self.model} with {self.battery_capacity} kWh battery"


# -----------------------------
# 3. Encapsulation Example
# -----------------------------
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew {amount}, New Balance: {self.__balance}"
        else:
            return "Insufficient funds!"

    def get_balance(self):
        return self.__balance


# -----------------------------
# 4. Polymorphism Example
# -----------------------------
class Bird:
    def make_sound(self):
        return "Some generic bird sound"

class Sparrow(Bird):
    def make_sound(self):
        return "Chirp Chirp!"

class Crow(Bird):
    def make_sound(self):
        return "Caw Caw!"


# -----------------------------
# 5. Math Helper Functions
# -----------------------------
def math_helper():
    while True:
        print("\n=== Math Helper ===")
        print("1. Square Root")
        print("2. Power")
        print("3. Factorial")
        print("4. Trigonometry (sin, cos, tan)")
        print("5. Back to Main Menu")

        choice = input("Enter choice: ")

        if choice == "1":
            num = float(input("Enter number: "))
            print("Square root:", math.sqrt(num))

        elif choice == "2":
            base = float(input("Enter base: "))
            exp = float(input("Enter exponent: "))
            print(f"{base}^{exp} =", math.pow(base, exp))

        elif choice == "3":
            num = int(input("Enter number: "))
            print(f"{num}! =", math.factorial(num))

        elif choice == "4":
            angle_deg = float(input("Enter angle in degrees: "))
            angle_rad = math.radians(angle_deg)
            print("sin:", math.sin(angle_rad))
            print("cos:", math.cos(angle_rad))
            print("tan:", math.tan(angle_rad))

        elif choice == "5":
            break
        else:
            print("Invalid choice!")


# -----------------------------
# 6. Date Calculator
# -----------------------------
def date_calculator():
    print("\n=== Date Calculator ===")
    today = date.today()
    print("Today's date:", today)
    days = int(input("Enter number of days to add/subtract: "))
    new_date = today + timedelta(days=days)
    print("New date:", new_date)


# -----------------------------
# 7. JSON Manager
# -----------------------------
def json_manager():
    data = {}
    while True:
        print("\n=== JSON Manager ===")
        print("1. Add entry")
        print("2. Show JSON")
        print("3. Save to file")
        print("4. Load from file")
        print("5. Back to Main Menu")

        choice = input("Enter choice: ")

        if choice == "1":
            key = input("Enter key: ")
            value = input("Enter value: ")
            data[key] = value
            print("‚úÖ Entry added.")

        elif choice == "2":
            print(json.dumps(data, indent=4))

        elif choice == "3":
            with open("data.json", "w") as f:
                json.dump(data, f, indent=4)
            print("‚úÖ Data saved to data.json")

        elif choice == "4":
            try:
                with open("data.json", "r") as f:
                    data = json.load(f)
                print("‚úÖ Data loaded from data.json")
            except FileNotFoundError:
                print("No file found!")

        elif choice == "5":
            break

        else:
            print("Invalid choice!")


# -----------------------------
# 8. RegEx Tester
# -----------------------------
def regex_tester():
    print("\n=== RegEx Tester ===")
    while True:
        print("\nOptions:")
        print("1. Search pattern")
        print("2. Find all matches")
        print("3. Replace pattern")
        print("4. Back to Main Menu")

        choice = input("Enter choice: ")

        if choice == "1":
            text = input("Enter text: ")
            pattern = input("Enter pattern: ")
            match = re.search(pattern, text)
            if match:
                print("Match found:", match.group())
            else:
                print("No match found.")

        elif choice == "2":
            text = input("Enter text: ")
            pattern = input("Enter pattern: ")
            matches = re.findall(pattern, text)
            print("Matches:", matches)

        elif choice == "3":
            text = input("Enter text: ")
            pattern = input("Enter pattern: ")
            replacement = input("Enter replacement text: ")
            new_text = re.sub(pattern, replacement, text)
            print("Result:", new_text)

        elif choice == "4":
            break

        else:
            print("Invalid choice!")


# -----------------------------
# 9. String Formatting Demo
# -----------------------------
def string_formatter():
    print("\n=== Python String Formatter ===")
    while True:
        print("\nOptions:")
        print("1. f-string")
        print("2. str.format()")
        print("3. % operator")
        print("4. Back to Main Menu")

        choice = input("Enter choice: ")

        if choice in ["1", "2", "3"]:
            name = input("Enter your name: ")
            age = input("Enter your age: ")

            if choice == "1":
                print(f"Hello, {name}! You are {age} years old.")
            elif choice == "2":
                print("Hello, {}! You are {} years old.".format(name, age))
            elif choice == "3":
                print("Hello, %s! You are %s years old." % (name, age))

        elif choice == "4":
            break

        else:
            print("Invalid choice!")


# -----------------------------
# Main Menu
# -----------------------------
def main_menu():
    while True:
        print("\n=== Python OOP & Concepts Menu ===")
        print("1. Class & Object")
        print("2. Inheritance")
        print("3. Encapsulation")
        print("4. Polymorphism")
        print("5. Math Helper")
        print("6. Date Calculator")
        print("7. JSON Manager")
        print("8. RegEx Tester")
        print("9. String Formatting")
        print("10. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            car = Car("Toyota", "Corolla")
            print(car.show_info())

        elif choice == "2":
            tesla = ElectricCar("Tesla", "Model 3", 75)
            print(tesla.show_info())

        elif choice == "3":
            account = BankAccount("Alice", 1000)
            print("Initial Balance:", account.get_balance())
            account.deposit(500)
            print("After Deposit:", account.get_balance())
            print(account.withdraw(300))
            print(account.withdraw(2000))

        elif choice == "4":
            birds = [Bird(), Sparrow(), Crow()]
            for b in birds:
                print(f"{b.__class__.__name__} ‚Üí {b.make_sound()}")

        elif choice == "5":
            math_helper()

        elif choice == "6":
            date_calculator()

        elif choice == "7":
            json_manager()

        elif choice == "8":
            regex_tester()

        elif choice == "9":
            string_formatter()

        elif choice == "10":
            print("Exiting... Goodbye!")
            break

        else:
            print("Invalid choice!")


# Run Program
if __name__ == "__main__":
    main_menu()

‚úÖ What This Updated Program Includes Now:

Class & Object ‚Üí Create and display a Car.

Inheritance ‚Üí ElectricCar inherits from Car.

Encapsulation ‚Üí BankAccount with private balance.

Polymorphism ‚Üí Bird, Sparrow, Crow with method overriding.

Math Helper ‚Üí Square root, power, factorial, trigonometry.

Date Calculator ‚Üí Add/subtract days from today.

JSON Manager ‚Üí Add entries, save/load JSON files.

RegEx Tester ‚Üí Search, find all, replace patterns.

String Formatting ‚Üí f-strings, .format(), % operator.
```

```python
Let‚Äôs go step by step on Python Virtual Environments.

üîπ What is a Virtual Environment?

A virtual environment (venv) is an isolated Python environment that allows you to:

Install packages without affecting the system Python

Manage different project dependencies separately

This is essential for Python projects to avoid version conflicts.

üîπ Creating a Virtual Environment

Open Command Prompt (Windows) or Terminal (Mac/Linux)

Navigate to your project folder:

cd path_to_your_project


Create the virtual environment

python -m venv myenv


myenv is the name of the virtual environment folder.

üîπ Activating the Virtual Environment

Windows:

myenv\Scripts\activate


Mac/Linux:

source myenv/bin/activate


‚úÖ After activation, you‚Äôll see (myenv) in your terminal.

üîπ Deactivating the Virtual Environment
deactivate

üîπ Installing Packages in Virtual Environment

Once activated, you can use pip to install packages locally in this environment:

pip install requests
pip list


These packages are isolated from other projects.

üîπ Deleting a Virtual Environment

Just delete the myenv folder.

No impact on your system Python.

üîπ Tiny Interactive Program Idea

Since Virtual Environments are managed via command line, Python itself cannot fully automate creating/activating them inside a running script.

But, you can include instructions and helper messages in your main OOP menu program:

def virtual_env_helper():
    print("=== Python Virtual Environment Helper ===")
    print("Steps to create a virtual environment:")
    print("1. Navigate to your project folder in terminal/command prompt.")
    print("2. Run: python -m venv myenv")
    print("3. Activate the environment:")
    print("   Windows: myenv\\Scripts\\activate")
    print("   Mac/Linux: source myenv/bin/activate")
    print("4. Install packages using pip.")
    print("5. Deactivate using: deactivate")
    print("6. Delete by removing the folder myenv if needed.")
```

```python
#Python File Handling
Let‚Äôs go step by step on Python File Handling.

üîπ Python File Handling Basics

Python allows us to create, read, update, and delete files.

when try to execute the code you have the example.txt in your drive

1. Opening a File

Use the open() function.

f = open("example.txt", "r")   # open file in read mode


"r" ‚Üí read (default)

"w" ‚Üí write (overwrites file)

"a" ‚Üí append (adds to file)

"x" ‚Üí create (fails if file exists)

"b" ‚Üí binary mode (for images, PDFs, etc.)

"t" ‚Üí text mode (default)

2. Reading Files
f = open("example.txt", "r")

print(f.read())         # read entire file
print(f.readline())     # read one line
print(f.readlines())    # read all lines into a list

f.close()

3. Writing Files
f = open("example.txt", "w")
f.write("Hello, World!\n")
f.write("Python file handling is easy.")
f.close()


‚ö†Ô∏è "w" overwrites the file.

4. Appending to Files
f = open("example.txt", "a")
f.write("\nAdding a new line.")
f.close()

5. Using with Statement (Best Practice)
with open("example.txt", "r") as f:
    content = f.read()
    print(content)


‚úÖ No need to call f.close(), it closes automatically.

6. Deleting Files
import os

if os.path.exists("example.txt"):
    os.remove("example.txt")
else:
    print("File does not exist")

7. File Handling Example Program
def file_handling_demo():
    print("=== File Handling Demo ===")

    # 1. Write to file
    with open("demo.txt", "w") as f:
        f.write("First line.\n")
        f.write("Second line.\n")

    # 2. Append
    with open("demo.txt", "a") as f:
        f.write("Appended line.\n")

    # 3. Read file
    with open("demo.txt", "r") as f:
        print("File Content:")
        print(f.read())


```

```python
#let‚Äôs go through the NumPy module step by step.

üîπ What is NumPy?

NumPy (Numerical Python) is a powerful Python library for:

Numerical computations

Handling large datasets

Scientific computing

Provides the ndarray object (multi-dimensional array) that is much faster than Python lists.

üîπ Installation
pip install numpy

üîπ Importing NumPy
import numpy as np

üîπ Creating Arrays
import numpy as np

# 1D array
arr1 = np.array([1, 2, 3, 4])
print(arr1)

# 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)

üîπ Useful Array Functions
a = np.zeros((2,3))     # 2x3 array of zeros
b = np.ones((3,3))      # 3x3 array of ones
c = np.eye(3)           # 3x3 identity matrix
d = np.arange(0,10,2)   # [0 2 4 6 8]
e = np.linspace(0,1,5)  # [0. 0.25 0.5 0.75 1.]

üîπ Array Operations
x = np.array([10, 20, 30, 40])
y = np.array([1, 2, 3, 4])

print(x + y)   # [11 22 33 44]
print(x - y)   # [ 9 18 27 36]
print(x * y)   # [10 40 90 160]
print(x / y)   # [10. 10. 10. 10.]

üîπ Matrix Operations
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(np.dot(A, B))  # Matrix multiplication
print(np.transpose(A))  # Transpose

üîπ Statistics with NumPy
data = np.array([1, 2, 3, 4, 5])

print(np.mean(data))  # 3.0
print(np.median(data)) # 3.0
print(np.std(data))   # 1.414...

üîπ Random Numbers
print(np.random.rand(3))     # 3 random numbers [0,1)
print(np.random.randint(1,10,5))  # 5 random ints between 1 and 9


‚úÖ NumPy = foundation for Data Science, AI, Machine Learning.
It is heavily used in Pandas, Scikit-learn, TensorFlow, PyTorch and more.
```

```python
Pandas is one of the most powerful and widely used Python libraries for data analysis and manipulation.
It‚Äôs built on top of NumPy and provides flexible data structures like:

Series ‚Üí 1D labeled array (like a column in Excel/SQL).

DataFrame ‚Üí 2D table (like an Excel sheet or SQL table).

üîπ Installing Pandas
pip install pandas

üîπ Basic Pandas Demo

Here‚Äôs a small script to show common operations:

import pandas as pd

# 1. Create a DataFrame (like a small table)
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Score": [85, 90, 95, 88]
}
df = pd.DataFrame(data)

print("üìå Original DataFrame:")
print(df)

# 2. Accessing columns
print("\nüîπ Ages column:")
print(df["Age"])

# 3. Filtering rows
print("\nüîπ Students older than 30:")
print(df[df["Age"] > 30])

# 4. Adding a new column
df["Passed"] = df["Score"] > 90
print("\nüîπ DataFrame after adding Passed column:")
print(df)

# 5. Basic statistics
print("\nüîπ Average Score:", df["Score"].mean())
print("üîπ Max Age:", df["Age"].max())

# 6. Saving to CSV
df.to_csv("students.csv", index=False)
print("\n‚úÖ Data saved to students.csv")

üîπ Output Example
üìå Original DataFrame:
      Name  Age  Score
0    Alice   25     85
1      Bob   30     90
2  Charlie   35     95
3    David   40     88

üîπ Ages column:
0    25
1    30
2    35
3    40
Name: Age, dtype: int64

üîπ Students older than 30:
      Name  Age  Score
2  Charlie   35     95
3    David   40     88

üîπ DataFrame after adding Passed column:
      Name  Age  Score  Passed
0    Alice   25     85   False
1      Bob   30     90   False
2  Charlie   35     95    True
3    David   40     88   False

üîπ Average Score: 89.5
üîπ Max Age: 40
‚úÖ Data saved to students.csv
```

```python
Let‚Äôs dive into SciPy, one of the most powerful Python libraries for scientific and technical computing üöÄ

üîπ What is SciPy?

SciPy stands for Scientific Python.

It is built on NumPy and provides additional functionality for:

Linear algebra

Optimization

Statistics

Signal & image processing

Integration & differentiation

Scientific computations

You usually import it as:

import scipy
from scipy import linalg, optimize, stats, integrate

üîπ Key SciPy Sub-packages

scipy.linalg ‚Üí Linear Algebra (like NumPy but more advanced)

scipy.optimize ‚Üí Functions for optimization/minimization

scipy.stats ‚Üí Statistical functions & probability distributions

scipy.integrate ‚Üí Numerical integration & solving ODEs

scipy.fft ‚Üí Fast Fourier Transform (signal processing)

scipy.sparse ‚Üí Sparse matrices & linear algebra

üîπ Mini Examples
1. Linear Algebra
from scipy import linalg
import numpy as np

# Solve Ax = b
A = np.array([[3, 2], [1, 4]])
b = np.array([7, 10])

x = linalg.solve(A, b)
print("Solution:", x)

2. Optimization
from scipy import optimize

# Minimize f(x) = x^2 + 10*sin(x)
func = lambda x: x**2 + 10*np.sin(x)
result = optimize.minimize(func, x0=0)  # start guess = 0
print("Minimum value found at x =", result.x)

3. Statistics
from scipy import stats

data = [1, 2, 3, 4, 5, 5, 6, 7]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)
print("Mean:", mean, "Median:", median, "Mode:", mode.mode[0])

4. Integration
from scipy import integrate

# ‚à´ sin(x) dx from 0 to œÄ
result, error = integrate.quad(lambda x: np.sin(x), 0, np.pi)
print("Integral result:", result)

5. Fourier Transform
from scipy import fft
import numpy as np

x = np.array([1, 2, 3, 4])
y = fft.fft(x)
print("FFT:", y)


‚úÖ In short:

NumPy ‚Üí Arrays & basic math

SciPy ‚Üí Advanced scientific computation
```

```python
Django is one of the most popular Python web frameworks for building scalable web applications quickly. Let‚Äôs go step by step with a beginner-friendly Django tutorial.

üîπ 1. Install Django

First, create a virtual environment (recommended) and install Django.

# Create virtual environment
python -m venv myenv

# Activate environment
# Windows:
myenv\Scripts\activate
# Mac/Linux:
source myenv/bin/activate

# Install Django
pip install django


Check if installed:

django-admin --version

üîπ 2. Create a Django Project
django-admin startproject mysite
cd mysite


The folder will look like this:

mysite/
    manage.py
    mysite/
        __init__.py
        settings.py
        urls.py
        asgi.py
        wsgi.py

üîπ 3. Run Development Server
python manage.py runserver


Visit: üëâ http://127.0.0.1:8000
You should see Django‚Äôs ‚ÄúIt worked!‚Äù page ‚úÖ

üîπ 4. Create an App

Inside the project:

python manage.py startapp myapp


Now you have:

myapp/
    admin.py
    apps.py
    models.py
    views.py
    urls.py   (we will create this file)

üîπ 5. Add App to Project

In mysite/settings.py, add "myapp" inside INSTALLED_APPS.

INSTALLED_APPS = [
    'django.contrib.admin',
    'django.contrib.auth',
    'django.contrib.contenttypes',
    'django.contrib.sessions',
    'django.contrib.messages',
    'django.contrib.staticfiles',
    'myapp',  # add this
]

üîπ 6. Create a View

Open myapp/views.py and add:

from django.http import HttpResponse

def home(request):
    return HttpResponse("Hello, Django! üéâ This is my first view.")

üîπ 7. Map URLs

Create myapp/urls.py:

from django.urls import path
from . import views

urlpatterns = [
    path('', views.home, name="home"),
]


Then include it inside mysite/urls.py:

from django.contrib import admin
from django.urls import path, include

urlpatterns = [
    path('admin/', admin.site.urls),
    path('', include('myapp.urls')),  # link app urls
]


Now go to üëâ http://127.0.0.1:8000/
You‚Äôll see your message: Hello, Django! üéâ

üîπ 8. Use Django Templates

Instead of raw text, let‚Äôs use HTML.

Create a folder myapp/templates/

Inside it, create home.html:

<!DOCTYPE html>
<html>
<head>
    <title>Django Demo</title>
</head>
<body>
    <h1>Welcome to Django üöÄ</h1>
    <p>This is a template example.</p>
</body>
</html>


Modify views.py:

from django.shortcuts import render

def home(request):
    return render(request, "home.html")

üîπ 9. Django Models (Database)

Inside myapp/models.py:

from django.db import models

class Student(models.Model):
    name = models.CharField(max_length=100)
    grade = models.IntegerField()

    def __str__(self):
        return self.name


Run migrations:

python manage.py makemigrations
python manage.py migrate

üîπ 10. Admin Panel

Create a superuser:

python manage.py createsuperuser


Register model in myapp/admin.py:

from django.contrib import admin
from .models import Student

admin.site.register(Student)


Now visit üëâ http://127.0.0.1:8000/admin and log in. You‚Äôll see Student table to add/edit records.

‚úÖ That‚Äôs your first Django app! You now have:

Views (logic)

Templates (HTML)

Models (database)

Admin Panel
```

```python
# Python
#Let‚Äôs go through Matplotlib step by step.

#üîπ What is Matplotlib?

#Matplotlib is the most popular data visualization library in Python.

#It allows you to create:

#Line charts

#Bar charts

#Histograms

#Pie charts

#Scatter plots


#3Custom visualizations


#It‚Äôs often used with NumPy and Pandas for data analysis.

#üîπ Installation
#%pip install matplotlib

#üîπ Basic Example
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)  # Line plot
plt.title("Simple Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()


#‚úÖ Output ‚Üí A straight line showing the relationship between X and Y.

#üîπ Common Plot Types
#1. Line Chart
plt.plot([1,2,3],[2,4,6], marker="o")
plt.title("Line Chart")
plt.show()

#2. Bar Chart
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()

#3. Histogram
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins=20, color="orange", edgecolor="black")
plt.title("Histogram")
plt.show()

#4. Scatter Plot
x = [5,7,8,7,6,9,5,6,7,8]
y = [99,86,87,88,100,86,103,87,94,78]

plt.scatter(x, y, color="red")
plt.title("Scatter Plot")
plt.show()

#5. Pie Chart
sizes = [30, 20, 25, 25]
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()

#üîπ Customization

#Color: color="red"

#Marker: marker="o" for points

#Line style: linestyle="--"

#Multiple plots: Use plt.plot() multiple times

#üîπ Tiny Interactive Program: Plot Menu
import matplotlib.pyplot as plt
import numpy as np

def plot_menu():
    while True:
        print("\n=== Matplotlib Plot Menu ===")
        print("1. Line Chart")
        print("2. Bar Chart")
        print("3. Histogram")
        print("4. Scatter Plot")
        print("5. Pie Chart")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            x = [1,2,3,4,5]
            y = [2,4,6,8,10]
            plt.plot(x,y, marker="o")
            plt.title("Line Chart")
            plt.show()

        elif choice == "2":
            categories = ["A","B","C","D"]
            values = [10,20,15,25]
            plt.bar(categories, values)
            plt.title("Bar Chart")
            plt.show()

        elif choice == "3":
            data = np.random.randn(500)
            plt.hist(data, bins=20, color="orange", edgecolor="black")
            plt.title("Histogram")
            plt.show()

        elif choice == "4":
            x = [5,7,8,7,6,9,5,6,7,8]
            y = [99,86,87,88,100,86,103,87,94,78]
            plt.scatter(x,y, color="red")
            plt.title("Scatter Plot")
            plt.show()

        elif choice == "5":
            sizes = [30,20,25,25]
            labels = ["Apples","Bananas","Cherries","Dates"]
            plt.pie(sizes, labels=labels, autopct='%1.1f%%')
            plt.title("Pie Chart")
            plt.show()

        elif choice == "6":
            print("Exiting Plot Menu...")
            break

        else:
            print("Invalid choice!")

if __name__ == "__main__":
    plot_menu()


#‚úÖ With this program you can interactively choose different plot types.
```
